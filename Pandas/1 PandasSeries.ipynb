```python
# 1. Write a program to create an empty series.

pd.Series([])
```

```python
# 2. Write a Pandas program to add, subtract, multiple and divide two Pandas Series.

s1 = pd.Series([2,4,6,8,10])
s2 = pd.Series([4,8,12,16,20])

print("Addition of two series\n",s1 + s2)
print("Subtraction of two series\n",s1 - s2)
print("Multiplication of two series\n",s1 * s2)
print("Division of two series\n",s1 / s2)
```

```python
# 3. Write a Pandas program to compare the elements of the two Pandas Series.
# Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 10]

s1 = pd.Series([2,4,6,8,10])
s2 = pd.Series([1,3,5,7,9])

print("Comparison of two series(>)\n",s1>s2)
print("Comparison of two series(<)\n",s1<s2)
print("Comparison of two series(=)\n",s1==s2)
```

```python
# 5. Write a function to change the data type of given a column or a Series. Function takes series and data type as input, 
# returns the converted series

def changeDataType(df,datatype):
    df = pd.to_numeric(df,downcast = datatype,errors="coerce")
    return df

batsman_runs = pd.read_csv("batsman_runs_series.csv")
print(changeDataType(batsman_runs["batsman_run"],"float"))
```

```python
# 6. Find top 10 most run getter from the series.

batsman_runs.sort_values("batsman_run",ascending = False).head(10)
```

```python
# 7. No of players having runs above 3000

batsman_runs[batsman_runs["batsman_run"] > 3000].shape[0]
```

```python
# 8. No of players having runs above mean value?

mean_runs = batsman_runs["batsman_run"].mean()
batsman_runs[batsman_runs["batsman_run"] > mean_runs].shape[0]
```

```python
# # 9.
# i. Read `items.csv` making `item_name` as index.
# ii. Show no of nan values
# ii. Item price is given in $, so convert it to rupees without currency symbol.
# iii. Make data type of newly made series as float.
# iv. Fill nan with mean of the series

def rupee(x):
    try:
        y = x[1:]
    except:
        y = x
    return float(y) * 80

items = pd.read_csv("items.csv",index_col = "item_name")
print("Number of Nan Values : ",items.isna().sum())
items["item_price"] = items["item_price"].apply(rupee)
print(items)
print("*" * 50)
items["item_price"] = items.fillna(items.mean())
print(items)
```

```python
# 10.
# i. Find mean price
# ii. Find 30th and 6th percentile value
# iii. Plot Histogram on price with bin size 50
# iv. No of items price lies between [1000 to 2000]

print("Mean Price : ",items["item_price"].mean())
print("30th Percentile : ",items["item_price"].quantile(q = 0.3))
print("6th Percentile : ",items["item_price"].quantile(q = 0.06))
items["item_price"].plot(kind = "hist",bins = 50)
print("No of items price lies between 1000 to 2000 is ",items[items["item_price"].between(1000,2000)].shape[0])
```
