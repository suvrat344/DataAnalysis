# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Plots

```python
x = [2,5,10]
y = [1,6,8]

plt.plot(x,y,'ko--')
plt.show()
```

```python
x  = [0,1,2,3,4,5]
y1 = [0,1,4,9,16,25]
y2 = [0,1,8,27,64,125]

plt.plot(x,y1,marker='o',linestyle='--',color='b',label='y = x^2')
plt.plot(x,y2,marker='o',linestyle='--',color='k',label='y = x^3')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Line Plot")

plt.legend()

plt.grid(True,c='k')

plt.show()
```

## Scatter Plot

```python
x = np.random.randint(100,size=100)
y = np.random.randint(100,size=100)


plt.scatter(x,y,marker='o',s=30)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Scatter Plot")

plt.show()
```

## Histogram

```python
x = np.random.normal(20,5,1000)

plt.hist(x,bins = 13)
plt.title("Histogram",fontsize = 25)

plt.show()
```

## Bar Chart

```python
categories = ['A','B','C','D','E']
values = np.random.randint(100,size=5)

plt.bar(categories,values,color='skyblue',width=0.7)

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title("Bar Chart")

plt.show()
```

## Pie Chart

```python
categories = ['A','B','C','D','E']
values = np.random.randint(100,size=5)

plt.pie(values,labels = categories,explode=[0,0.1,0,0,0],autopct='%.1f%%',startangle=90,shadow=True)

plt.title("Pie Chart")

plt.show()
```

## Box Plot

```python
x = np.random.normal(200,8,1000)

IQR = np.quantile(x,0.75) - np.quantile(x,0.25)

plt.boxplot(x)

plt.title("Box Plot")
plt.yticks([np.quantile(x,0.25) - 1.5 * IQR,np.quantile(x,0.25),np.median(x),np.quantile(x,.75),np.quantile(x,0.75)+1.5 * IQR])

plt.show()
```

## Subplotting

```python
x = np.arange(100)

p1 = np.sin(x)
p2 = np.cos(x)

plt.subplot(221)
plt.plot(p1)
plt.title("Sine Function")

plt.subplot(222)
plt.plot(p2)
plt.title("Cosine Function")

plt.show()
```

## 3D Plotting

```python
ax = plt.axes(projection="3d")

x = np.random.randint(100,size=100)
y = np.random.randint(100,size=100)
z = np.random.randint(100,size=100)

ax.scatter(x,y,z)

plt.show()
```

## Animating the plots

```python
%matplotlib notebook

n_tosses = 500

head_counts = 0
tosses = np.arange(1,n_tosses + 1)

fig,ax = plt.subplots()
ax.set_xlim(1,n_tosses)
ax.set_ylim(0,1)
line, = ax.plot([],[],color='blue')

ax.set_xlabel("Number of tosses")
ax.set_ylabel("Proportion of heads")
ax.set_title("Coin toss simulation")

proportion_heads = []

for i in range(1,n_tosses+1):
    toss = np.random.randint(0,2)
    
    if(toss == 1):
        head_counts += 1
    
    proportion_heads.append(head_counts/i)
    
    line.set_data(tosses[:i],proportion_heads)
    
    plt.pause(0.01)
    
plt.show()
```
