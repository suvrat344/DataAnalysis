```python
# Add a label to every axis and add a proper title for the charts. Also add proper labels if there are multiple representations.
# Then, you can customize it as your wish.
# 1 .Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China

df = pd.read_csv("MostPollutedCities.csv")

df1 = df[df["Country"] == "Iran"].groupby("Year")["PM2.5"].sum().reset_index()
df2 = df[df["Country"] == "China"].groupby("Year")["PM2.5"].sum().reset_index()
plt.plot(df1["Year"],df1["PM2.5"],label="Iran",linestyle="dashed",linewidth=1)
plt.plot(df2["Year"],df2["PM2.5"],label="China",linestyle="dotted",linewidth=1)
plt.title("PM2.5 Level in China and Iran in Different Year")
plt.xlabel("Year")
plt.xticks(df['Year'].value_counts().index)
plt.ylabel("PM2.5")
plt.legend(loc="best")
plt.show()
```

```python
# 2. Draw a histogram of the column "PM10" of which the y-axis represents the probability (see the documentation how to draw 
# the probability).

plt.hist(df["PM10"],density=True,bins=50,alpha=0.6,color="green")
plt.title("Histogram of PM10")
plt.xlabel("Bins")
plt.ylabel("Probability")
plt.show()
```

```python
# 3. Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

df1 = df[df["Country"] == "Poland"]
df2 = df[df["Country"] == "Chile"]
plt.scatter(df1["PM2.5"],df1["PM10"],label="Poland",marker="+")
plt.scatter(df2["PM2.5"],df2["PM10"],label="Chile",marker="D")
plt.title("PM2.5 vs PM10 in Poland and Chile")
plt.xlabel("PM2.5")
plt.ylabel("PM10")
plt.legend()
plt.show()
```

```python
# 1 Make a subplots which have 2 plots.
# For the first chart, draw a scatter plot "Monitored Cap.(MW)" vs "Total Cap. Under Maintenance (MW)" of top 5 most frequent 
# power stations. Then draw the lines which indicate the average values of these two columns. Change the colors according to 
# the names of the Power Stations.
# For the second chart, draw a scatter plot "Monitored Cap. (MW)" vs "Actual(MU)" of the top 5 most frequent power stations. 
# Also draw the lines which indicates the average values of these two columns. Change the colors according to the names of the 
# Power Stations.

power = pd.read_csv("PowerGeneration.csv")

freq = power["Power Station"].value_counts().head().index.tolist()
temp_df = power[power["Power Station"].isin(freq)]
encoder = LabelEncoder()
temp_df['Power Station'] = encoder.fit_transform(temp_df['Power Station'])

fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(12,5))

ax[0].scatter(temp_df["Monitored Cap.(MW)"],temp_df["Total Cap. Under Maintenace (MW)"],c=temp_df["Power Station"])
ax[0].axhline(temp_df["Monitored Cap.(MW)"].mean(),c="red")
ax[0].axvline(temp_df["Total Cap. Under Maintenace (MW)"].mean(),color="black")
ax[0].set_title("Monitored Cap.(MW) vs Total Cap. Under Maintenace (MW)")
ax[0].set_xlabel("Monitored Cap.(MW)")
ax[0].set_ylabel("Total Cap. Under Maintenance (MW)")

ax[1].scatter(temp_df["Monitored Cap.(MW)"],temp_df["Actual(MU)"],c=temp_df["Power Station"])
ax[1].axhline(temp_df["Actual(MU)"].mean(),color="red")
ax[1].axvline(temp_df["Monitored Cap.(MW)"].mean(),c="black")
ax[1].set_title("Monitored Cap.(MW) vs Actual(MU)")
ax[1].set_xlabel("Monitored Cap.(MW)")
ax[1].set_ylabel("Actual(MU)")
plt.show()
```

```python
# Draw a 3D Scatter plot between "Monitored Cap.(MW)", "Total Cap. Under Maintenace (MW)" and "Forced Maintanence(MW)"

fig = plt.figure(figsize=(15,15))
ax = plt.subplot(projection="3d")
ax.scatter3D(power["Monitored Cap.(MW)"],power["Total Cap. Under Maintenace (MW)"],power["Forced Maintanence(MW)"],marker="+")
ax.set_title("Monitored Cap.(MW) vs Total Cap. Under Maintenace (MW) vs Forced Maintanence(MW)")
ax.set_xlabel("Monitored Cap.(MW)")
ax.set_ylabel("Total Cap. Under Maintenace (MW)")
ax.set_zlabel("Forced Maintanence(MW)")
plt.show()
```

$$z = sin(\sqrt{x^2 + y^2})$$

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = np.sin(np.sqrt(xx**2 + yy**2))

fig = plt.figure(figsize=(8,8))
ax = plt.subplot(projection="3d")
p = ax.plot_surface(xx,yy,z,cmap="viridis")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
fig.colorbar(p)
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap="viridis")
ax.set_xlabel("x")
ax.set_ylabel("y")
fig.colorbar(p)
plt.show()
```
